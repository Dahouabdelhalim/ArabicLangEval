```python
import pandas as pd

# Load your dataset
df = pd.read_csv("/content/WSD_Arabic_Dataset.xlsx - Sheet1.csv", sep=',')

# Remove duplicated rows based on ID + Word + Meaning
df_clean = df.drop_duplicates(subset=["ID", "Word", "Meaning"], keep="first")

# Save the cleaned file
df_clean.to_csv("WSD_Arabic_Dataset_cleaned_file.csv", index=False)

print("Original rows:", len(df))
print("Cleaned rows:", len(df_clean))
```
