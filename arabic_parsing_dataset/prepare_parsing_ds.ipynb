```python
# Prepare data code by combining MASAQ dataset with Quran dataset to get the full verse text

import pandas as pd

# Load the CSV MASQA dataset
MASAQ_df = pd.read_csv("MASAQ.csv", sep=",",low_memory=False)  # adjust delimiter if needed

print(len(MASAQ_df))

#load quran_dataset 
quran_df = pd.read_json("quran_dataset/data/holy_quran.json")

#get ayah full text
def get_ayah(surah_no,ayah_no):
    return quran_df["data"]["surahs"][surah_no]["ayahs"][ayah_no]["text"]

full_verse_col = []
for index, row in MASAQ_df.iterrows():
    full_verse = get_ayah(row['Sura_No']-1,row['Verse_No']-1)  #get verse text
    full_verse_col.append(full_verse)

print(len(full_verse_col))
MASAQ_df['Verse_text'] = full_verse_col


MASAQ_df.to_csv("combined_dataset.csv", index=False)
```

Formulate the parsing of a word

```python
import pandas as pd

df = pd.read_csv("combined_dataset.csv")

parsed_words_lst= []

for idx, row in df.iterrows():
    syntactic = row["Syntactic_Role"]
    case_mood = row["Case_Mood"]
    marker = row["Case_Mood_Marker"]
    if pd.isna(syntactic) or str(syntactic).strip() == "":
        continue
    # If "Case_Mood" is empty, add "" after it; otherwise, add "" to the beginning of "Case_Mood_Marker"
    if pd.isna(case_mood) or str(case_mood).strip() == "مبني":
        marker = "على " + str(marker)
    else:
        marker = "ب" + str(marker)
    word_parsing = f"{syntactic} {case_mood} {marker}"
    parsed_words_lst.append([row["ID"], row["Sura_No"], row["Verse_No"], row["Word_No"], row["Word"], row["Segmented_Word"], word_parsing,row["Verse_text"]])

#save to CSV
parsed_df = pd.DataFrame(parsed_words_lst, columns=["ID", "Sura_No", "Verse_No", "Word_No", "Word", "Segmented_Word", "Syntactic_Parsing", "Verse_text"])
parsed_df.to_csv("parsed_words_dataset.csv", index=False)
```
